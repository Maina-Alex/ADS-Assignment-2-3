```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.set_index('EmployeeNumber')
```

```python

df['Attrition'].replace({"Yes":1,"No":0},inplace=True)
```

```python
df
```

```python
df.columns
```

```python

df.groupby('Department').mean()['Attrition'].plot(kind='bar',color=['Blue','Orange','Green'])
plt.title("Attrition by Department")

```

```python
df.groupby('JobLevel').mean()['Attrition'].plot(kind='bar',color=['Blue','Orange','Green'])
plt.title("Attrition by JobLevel")
```

```python

df.groupby('Education').mean()['Attrition'].plot(kind='bar',color=['Blue','Orange','Green'])
plt.title("Attrition by Education")
```

```python
df.groupby('Education').mean()['MonthlyIncome'].plot(kind='bar',color=['Blue','Orange','Green'])
plt.title("Education by MonthlyIncome")
```

```python
plt.subplots(12,figsize=[15,6])
plt.subplot(121)
sns.boxplot(x='Attrition',y='MonthlyIncome',data=df)
plt.title("Attrition Rate by Monthly Income")
plt.subplot(122)
sns.distplot(df['MonthlyIncome'])
plt.xlim(0,1750)
sns.despine()
```

```python
plt.figure(figsize=([15,25]))
plt.subplots(15,figsize=[15,6])
plt.subplot(121)
sns.boxplot(x='Attrition',y='DistanceFromHome',data=df)
plt.title("Attrition Rate by Distance From Home")
plt.subplot(122)
sns.distplot(df['DistanceFromHome'])
plt.xlim(0,30)
sns.despine()
```
